```python
import numpy as np
from scipy import linalg
import matplotlib.pyplot as plt
```

```python
ts = np.linspace(0, 2*np.pi)
xs = np.cos(ts)
ys = np.sin(ts)
```

```python
# Let A be any matrix
A = np.array([
    [1.5, 2],
    [-0.2, 0.3],
])

# Compute image A [x_i,y_i] for each point x_i, y_i in (xs, ys)
Ax = np.array([A @ [x, y] for x, y in zip(xs, ys)])
```

```python
# Compute SVD
U, S, Vt = linalg.svd(A)
V = Vt.transpose()
```

```python
fig, axes = plt.subplots(ncols=2, figsize=(12,6))

# Circle
axes[0].plot(x, y)
axes[0].set(xlim=(-3,3), ylim=(-3,3))
axes[0].set_aspect('equal')
axes[0].set_title('Unit circle')

# Image of circle
axes[1].plot(Ax[:, 0], Ax[:, 1])
axes[1].set(xlim=(-3,3), ylim=(-3,3))
axes[1].set_aspect('equal')
axes[1].set_title('Transformed unit circle')

# Circle vectors
axes[0].quiver(0, 0, V[0,0], V[1, 0], units='x', scale=1, color='black')
axes[0].quiver(0, 0, V[0,1], V[1, 1], units='x', scale=1, color='red')

# Image vectors
axes[1].quiver(0, 0, S[0]*U[0,0], S[0]*U[1, 0], units='x', scale=1, color='black')
axes[1].quiver(0, 0, S[1]*U[0,1], S[1]*U[1, 1], units='x', scale=1, color='red')
```
