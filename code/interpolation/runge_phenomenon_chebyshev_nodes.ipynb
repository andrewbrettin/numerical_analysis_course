```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.interpolate import lagrange
from scipy.special import roots_chebyt
```

```python
n = 15
xs = np.linspace(-1,1,n)
xs_fine = np.linspace(-1,1,10*n)

# Function to interpolate: 
f = lambda x : 1/(1 + 25 * x**2)
# Interpolating polynomial
p = lagrange(xs, f(xs))

plt.plot(xs_fine, f(xs_fine), label='Runge function')
plt.plot(xs, f(xs), 'mo', label='Equidistant nodes')
plt.plot(xs_fine, p(xs_fine), label=f'Degree {n} interpolating polynomial')
plt.legend(bbox_to_anchor=(1,0.5))
```

Using chebyshev nodes instead of equally spaced nodes:

```python
from numpy.polynomial.chebyshev import chebpts1, chebpts2

n = 15
xs = roots_chebyt(n)[0]
xs_fine = np.linspace(-1,1,10*n)

f = lambda x : 1/(1 + 25 * x**2)
p = lagrange(xs, f(xs))

plt.plot(xs_fine, f(xs_fine), label='Runge function')
plt.plot(xs, f(xs), 'mo', label='Chebyshev nodes')
plt.plot(xs_fine, p(xs_fine), label=f'Degree {n} Chebyshev polynomial')
plt.legend(bbox_to_anchor=(1,0.5))
```
